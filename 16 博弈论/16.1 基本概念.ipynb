### 16.1.1 博弈论的定义
博弈论是描述、分析多人决策行为的一种决策理论，是多个主体在相互影响下的多元决策，决策的均衡结果取决于双方或多方的决策

### 16.1.2 博弈论中的经典案例
#### 1. 囚徒困境
反映了个体理性行为与集体理性行为之间的矛盾与冲突
#### 2. 智猪博弈
纳什均衡，指的是在给定一方采取某种策略的条件下，另一方所采取的最佳策略
#### 3. 斗鸡博弈
在该种博弈中，非理性、非理智的形象塑造往往是一种可选择的策略运用

### 16.1.3 博弈的一般概念
#### 1. 局中人 (Player)
博弈中有权决定自己行动方案的博弈参加者，$I={1,2,\cdots,n}$
#### 2. 策略集 (Strategies)
博弈中可供局中人选择的实际可行的完整行动方案的集合，记局中人 $i$ 的策略集为 $S_i$
#### 3. 赢得函数（支付函数）(Payoff Function)
博弈中，每一局中人所出策略形成的策略组为一个局势 $s=\left(s_1, s_2,\cdots,s_n\right)$，$n$ 为局中人的个数。在局势 $s$ 下，为每个局中人 $i$ 规定一个赢得值 $H_i(s)$，则 $H_i(s)$ 为定义在所有局势的集合 $S$ 上的函数，称此函数为局中人 $i$ 的赢得函数

> #### 博弈问题的分类
> 
> - 根据策略与时间的关系：静态博弈、动态博弈
> 
> - 根据局中人的数目：二人博弈、多人博弈（合作博弈、非合作博弈）
> - 根据各局中人的赢得函数的代数和是否为零：零和博弈、非零和博弈
> - 根据策略的概率特性：纯策略博弈、混合策略博弈
> - 根据局中人的策略集的策略数：有限策略博弈、无限策略博弈
> - 根据数学模型的不同：**矩阵博弈**、连续博弈、微分博弈、阵地博弈、随机博弈
