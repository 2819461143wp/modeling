### 3
多种拟合函数，根据残差方差选择拟合模型

```python
data = np.loadtxt('../../../hw/10第10章  回归分析习题解答/ti10_3.txt')
x, y = data[::2].flatten()[:-1], data[1::2].flatten()[:-1]
plt.scatter(x, y)
plt.show()
```

```python
n = len(x)

p1 = np.polyfit(x, y, 1)    # 线性 y = a1 + a2*x
yh1 = np.polyval(p1, x)
s1 = sum((y - yh1)**2/(n - 2))

p2 = np.polyfit(x, y, 2)    # 二次多项式 y = a1 + a2*x + a3*x**2
yh2 = np.polyval(p2, x)
s2 = sum((y - yh2)**2/(n - 3))

mat1 = np.vstack([np.ones(n), 1/x]).T
p3 = np.linalg.pinv(mat1) @ y   # 一支双曲线 y = a1 +　a2*(1/x)
yh3 = mat1 @ p3
s3 = sum((y - yh3)**2/(n - 2))

mat2 = np.vstack([np.ones(n), np.log(x)]).T
p4 = np.linalg.pinv(mat2) @ y  # 对数曲线 y = a1 + a2*lnx
yh4 = mat2 @ p4
s4 = sum((y - yh4)**2/(n-2))

print(s1, s2, s3, s4)
```

单支双曲线 $y=a_1+\dfrac{a_2}{x}$ 拟合效果最好

```python
xx = np.linspace(x.min(), x.max(), 100)
n = len(xx)
mat = np.vstack([np.ones(n), 1/xx]).T
yh = mat @ p3

fig, ax = plt.subplots()
ax.plot(xx, yh)
ax.scatter(x, y, c='r')
fig.show()
```

### 7
多元多项式 Logistic 拟合（分组），自变量估计，赔率计算

这里是线性拟合，可以使用更通用的矩阵乘法进行计算

```python
data = np.loadtxt('../../../hw/10第10章  回归分析习题解答/ti10_7.txt')
data = np.vstack([data[:,:4], data[:,4:]])
x1, x2, y, freq = data.T
test_num = 100
pos, neg = [], []
for i, yi in enumerate(y):
    if yi == 1:
        pos.append(freq[i])
        neg.append(test_num - freq[i])
    else:
        pos.append(test_num - freq[i])
        neg.append(freq[i])
y = np.vstack([pos, neg]).T
mod_dic = {'x1': x1, 'x2': x2, 'y': y}
mod = sm.formula.glm('y~x1*x2', mod_dic, family=sm.families.Binomial()).fit()
print(mod.summary())
```

```python
xx1 = np.linspace(1, 10, 100)
yy1 = mod.predict({'x1': xx1, 'x2': np.ones(len(xx1))})
yy2 = mod.predict({'x1': xx1, 'x2': np.zeros(len(xx1))})
p = np.array(pos)/test_num
x01 = [x1[i] for i in range(len(x2)) if x2[i]==1]
y01 = [p[i] for i in range(len(x2)) if x2[i]==1]
x02 = [x1[i] for i in range(len(x2)) if x2[i]==0]
y02 = [p[i] for i in range(len(x2)) if x2[i]==0]

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(xx1, yy1, label='加盖')
ax.plot(xx1, yy2, label='不加盖')
ax.scatter(x01, y01, marker='o', zorder=10, label="加盖")
ax.scatter(x02, y02, marker='s', zorder=10, label="不加盖")
ax.set_xlabel('水分')
ax.set_ylabel('预测发芽率')
from matplotlib.ticker import MultipleLocator
ax.xaxis.set_major_locator(MultipleLocator(1))
ax.legend(fontsize=8)
ax.grid()
fig.show()
```

```python
from scipy.optimize import root

fhalf = lambda x1, x2: mod.predict({'x1': x1, 'x2': x2}) - 0.5

print("加盖时发芽率为0.5的水分：", root(fhalf, 8, args=1)['x'][0])
print("不加盖时发芽率为0.5的水分：", root(fhalf, 4, args=0)['x'][0])
```

```python
fpred = lambda x1, x2: mod.predict({'x1': x1, 'x2': x2})[0]
odds6_1 = fpred(6, 1)/(1-fpred(6, 1))
odds6_0 = fpred(6, 0)/(1-fpred(6, 0))
print("水分为6时，若加盖发芽的赔率：", odds6_1)
print("水分为6时，若不加盖发芽的赔率：", odds6_0)
print("赔率比：", odds6_1/odds6_0)
```

### 8
线性和非线性方法估计参数

```python
data = np.loadtxt('../../../hw/10第10章  回归分析习题解答/ti10_8.txt')
x, y = data
# 1/y = a0 + a1*(1/x)
n = len(x)
R = np.vstack([np.ones(n), 1/x]).T
A = np.linalg.pinv(R) @ (1/y)
print("线性拟合参数值：", A)
```

```python
from scipy.optimize import least_squares

def func(u, x):
    u1, u2 = u
    return x/(u1*x + u2)

def err(u, x, y):
    return func(u, x) - y

u = np.random.rand(2)
uhat = least_squares(err, u, args=(x, y))['x']
print("线性拟合参数值：", uhat)
```
